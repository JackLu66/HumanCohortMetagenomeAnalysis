Calculation of microbiome-host phenotype associations

For microbial compositional data, species-level features were transformed using the centered log-ratio (CLR) method. Spearman correlation analyses were then performed between the transformed microbial profiles and host aging-related features. Similarly, for B. pseudocatenulatum, CLR transformation was applied to the microbial abundance data, followed by Spearman correlation analysis with host aging-related features. To correct for multiple testing, the Benjamini-Hochberg procedure was applied, with a false discovery rate (FDR, q value) < 0.05 considered statistically significant. Cytoscape software was used to generate correlation networks, and features from each omics dataset were manually clustered together for visualization.
