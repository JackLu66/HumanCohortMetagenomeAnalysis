To identify genes involved in 5-AVAB biosynthesis, a custom database specific to the 5-AVAB biosynthetic pathway was constructed. This database was curated based on 5-AVAB biosynthetic genes retrieved from UniProt. Subsequently, the extracted ORFs were subjected to BLASTx analysis (version 2.13.0) against the aforementioned custom database, with an E-value cutoff set at 1e-5. For further analysis, only those hits satisfying the stringent criteria of ≥ 70% sequence identity and ≥ 50% query coverage were retained.

```python
makeblastdb -in uniprotkb_5-AVAB biosynthetic genes.fasta \
-dbtype prot -parse_seqids -out /5-AVAB biosynthetic genes/index

conda activate /dellstorage06/quj_lab/luxiaoyong/00_software/miniconda3/envs/metag
dir_i=/MAG_Bp
cd ${dir_i}
samples=`ls . | sort -n`

cd /Result

for i in ${samples}
do
blastx -query ${dir_i}/${i} \
-db /5-AVAB biosynthetic genes/index \
-evalue 1e-6 -outfmt 6 -num_threads 6 -out Out_file_${i}
done
```
