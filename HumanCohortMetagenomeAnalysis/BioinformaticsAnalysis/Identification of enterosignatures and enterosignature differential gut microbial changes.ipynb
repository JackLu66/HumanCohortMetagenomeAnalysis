Identification of enterosignatures and enterosignature differential gut microbial changes

To comprehensively characterize gut microbial guilds with aging, we applied the Non-Negative Matrix Factorization (NMF)-based enterosignature framework described by Frioux et al. for our cohort analysis. Detailed procedures are as follows. NMF was performed using Scikit-Learn (version 0.24.1) with the multiplicative update solver, Kullback-Leibler divergence as the beta-loss function, random initialization (max iterations = 2000), and L1 regularization (α = 1.0) applied to both basis (genera-enterosignature weight) and mixture coefficient (enterosignature-sample abundance) matrices. Nine-fold bicross-validation was used to determine the optimal number of enterosignatures (k), with the gain in cosine similarity between the reconstructed and original abundance matrices as the key criterion. Cosine similarity was calculated to determine the optimal number of guilds. Define the normalized values of the guilds as the enterosignature scores. We further calculated the correlation between the enterosignature scores and age. Based on the enterosignatures, we further calculated the correlation between the diversity index of the microbial intestinal flora characteristics and age.
